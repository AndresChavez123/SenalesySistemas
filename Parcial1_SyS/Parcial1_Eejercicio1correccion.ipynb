**RANGO DE LA SEÑAL**

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear un vector de tiempo t desde 0 hasta 4π (dos periodos de la señal) con 1000 puntos
t = np.linspace(0, 4 * np.pi, 1000)

# Definir la señal x(t) compuesta por:
# - Un seno de frecuencia 7 rad/s, con desfase de π/2 y amplitud 20
# - Un coseno de frecuencia 5 rad/s y amplitud -3
# - Un coseno de frecuencia 10 rad/s y amplitud 2
x = 20 * np.sin(7 * t - np.pi / 2) - 3 * np.cos(5 * t) + 2 * np.cos(10 * t)

# Calcular el valor mínimo y máximo de la señal (rango de amplitud)
x_min = np.min(x)
x_max = np.max(x)

# Mostrar el rango de la señal en voltios
print(f"Rango de x(t): {x_min:.2f} V a {x_max:.2f} V")

# Graficar la señal x(t)
plt.figure(figsize=(10, 4))         # Crear una figura de tamaño ancho x alto
plt.plot(t, x, label='x(t)')        # Graficar la señal
plt.title('Señal original x(t)')    # Título del gráfico
plt.xlabel('Tiempo [s]')            # Etiqueta del eje x
plt.ylabel('Amplitud [V]')          # Etiqueta del eje y
plt.grid(True)                      # Agregar rejilla al gráfico
plt.legend()                        # Mostrar la leyenda
plt.show()                          # Mostrar el gráfico
```

**ACONDICIONAMIENETO DE LA SEÑAL**

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------------------------------------
# 1. Crear vector de tiempo para dos periodos completos
# -----------------------------------------------------------
t = np.linspace(0, 4 * np.pi, 1000)  # Tiempo desde 0 hasta 4π con 1000 puntos

# -----------------------------------------------------------
# 2. Definir la señal original x(t)
#    Compuesta por senos y cosenos de distintas frecuencias
# -----------------------------------------------------------
x = 20 * np.sin(7 * t - np.pi / 2) - 3 * np.cos(5 * t) + 2 * np.cos(10 * t)

# -----------------------------------------------------------
# 3. Calcular y mostrar el rango de la señal original
# -----------------------------------------------------------
x_min = np.min(x)
x_max = np.max(x)
print(f"Rango de x(t): {x_min:.2f} V a {x_max:.2f} V")

# -----------------------------------------------------------
# 4. Graficar la señal original
# -----------------------------------------------------------
plt.figure(figsize=(10, 4))
plt.plot(t, x, label='x(t)')
plt.title('Señal original x(t)')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [V]')
plt.grid(True)
plt.legend()
plt.show()

# -----------------------------------------------------------
# 5. Definir el rango del ADC (convertidor analógico-digital)
# -----------------------------------------------------------
Vmin = -3.3  # Voltaje mínimo del ADC
Vmax = 5.0   # Voltaje máximo del ADC

# -----------------------------------------------------------
# 6. Calcular ganancia (a) y offset (b) para acondicionar la señal
#    Formula: x_acond = a * x + b, tal que quede dentro del rango [Vmin, Vmax]
# -----------------------------------------------------------
a = (Vmax - Vmin) / (x_max - x_min)
b = Vmin - a * x_min

# -----------------------------------------------------------
# 7. Aplicar el acondicionamiento a la señal
# -----------------------------------------------------------
x_acond = a * x + b

# -----------------------------------------------------------
# 8. Mostrar los parámetros de acondicionamiento
# -----------------------------------------------------------
print(f"Ganancia (a): {a:.4f}")
print(f"Offset (b): {b:.4f}")
print(f"x_acond(t) ∈ [{np.min(x_acond):.2f}, {np.max(x_acond):.2f}] V")

# -----------------------------------------------------------
# 9. Graficar la señal acondicionada
# -----------------------------------------------------------
plt.figure(figsize=(10, 4))
plt.plot(t, x_acond, label='x_acond(t)', color='orange')
plt.title('Señal acondicionada')
plt.xlabel('Tiempo [s]')
plt.ylabel('Voltaje [V]')
plt.grid(True)
plt.legend()
plt.show()

```

```python
# Generamos por separado las tres gráficas de las señales: original, acondicionada y digitalizada

# Para graficar correctamente las señales acondicionada y digitalizada que tienen 2000 puntos,
# necesitamos un vector de tiempo con la misma cantidad de puntos.
# Vamos a recrear 't' con 2000 puntos, asumiendo que el acondicionamiento y digitalización
# se realizaron sobre una señal con esta cantidad de puntos, aunque el código anterior
# definía 't' con 1000. Si el acondicionamiento/digitalización se hizo con 1000 puntos,
# entonces x_cond y x_digital deberían tener 1000 puntos y el error estaría en su cálculo.
# Asumimos que el cálculo de x_cond y x_digital generó correctamente 2000 puntos y
# que el vector de tiempo para la última celda debe ser de 2000 puntos.
t_plot = np.linspace(0, 4 * np.pi, 2000)

# Señal original (Aunque la señal original 'x' se calculó con 1000 puntos,
# si queremos mostrarla junto a las otras, también necesitaría 2000 puntos.
# Para simplificar y evitar recalcular 'x', asumiremos que 'x' en este contexto
# representa la señal original muestreada a 2000 puntos, o que la gráfica
# de la señal original debería usar el 't' original de 1000 puntos si no se
# compara directamente en la misma figura con las otras dos.
# Dado que se pide graficar las tres por separado, usaremos el 't' original de 1000
# para la gráfica de la señal original y el nuevo 't_plot' de 2000
# para las otras dos, si es que x_cond y x_digital realmente tienen 2000 puntos.
# Si x_cond y x_digital deberían tener 1000 puntos, la corrección debería ser en
# la sección de ACONDICIONAMIENTO para que generen 1000 puntos.
# Basado en el traceback y las variables, x_cond y x_digital tienen 2000 puntos,
# así que ajustamos los tiempos para graficar esas dos.

# Señal original (usando el 't' original de 1000 puntos)
plt.figure(figsize=(10, 4))
# Corregir x_t a x
plt.plot(t, x, label="x(t)", color='blue')
plt.title("Señal original x(t)")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# --- Inicio de cambios ---
# Recalcular la señal acondicionada and digitalizada with 2000 points for plotting
t_full = np.linspace(0, 4 * np.pi, 2000)
x_full = 20 * np.sin(7 * t_full - np.pi / 2) - 3 * np.cos(5 * t_full) + 2 * np.cos(10 * t_full)
x_min_full = np.min(x_full)
x_max_full = np.max(x_full)
a_full = (Vmax - Vmin) / (x_max_full - x_min_full)
b_full = Vmin - a_full * x_min_full
x_acond_full = a_full * x_full + b_full

# Add digitalization (example with 5 bits)
n_bits = 5
levels = 2**n_bits
# Quantization step
q_step = (Vmax - Vmin) / (levels - 1)
# Quantize the signal
x_digital_full = np.round((x_acond_full - Vmin) / q_step) * q_step + Vmin
# Ensure values are within the ADC range after quantization
x_digital_full = np.clip(x_digital_full, Vmin, Vmax)
# --- Fin de cambios ---


# Señal acondicionada (usando t_plot con 2000 puntos)
plt.figure(figsize=(10, 4))
# Usar t_plot para el eje x y corregir x_cond a x_acond
plt.plot(t_plot, x_acond_full, label="x_acond(t)", color='orange') # Usamos x_acond_full
plt.title("Señal acondicionada al rango del ADC [-3.3V, 5V]")
plt.xlabel("Tiempo [s]")
plt.ylabel("Voltaje [V]")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Señal digitalizada (usando t_plot con 2000 puntos)
plt.figure(figsize=(10, 4))
# Usar t_plot para el eje x y usar la nueva variable x_digital_full
plt.plot(t_plot, x_digital_full, drawstyle='steps-post', label="x_digital(t)", color='green') # Usamos x_digital_full
plt.title("Señal digitalizada (5 bits - 32 niveles)")
plt.xlabel("Tiempo [s]")
plt.ylabel("Voltaje cuantizado [V]")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
