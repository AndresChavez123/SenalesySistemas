```python
from sympy import symbols, laplace_transform, exp, Heaviside, simplify
t, s, b = symbols('t s b', real=True)
f = exp(b*t) * Heaviside(t)
F = laplace_transform(f, t, s, noconds=True)
simplify(F)
```

```python
from sympy import cos, symbols
from sympy.abc import t, s
from sympy.integrals.transforms import laplace_transform
from sympy import exp, Heaviside, simplify

omega0 = symbols('omega0', real=True)

f = cos(omega0*t) * Heaviside(t)
F = laplace_transform(f, t, s, noconds=True)
simplify(F)
```
